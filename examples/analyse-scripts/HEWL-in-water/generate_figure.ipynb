```python
import os
import numpy as np
import MDAnalysis as mda

import matplotlib.pyplot as plt

all_files = ["T300K_ratio0.08", "T300K_ratio0.17", "T300K_ratio0.25", "T300K_ratio0.39", "T300K_ratio0.52", "T300K_ratio0.73"]

for file in all_files:

    ratio = np.float32(file[-4:])

    data = np.load(file+".npy", allow_pickle=True)
    f = data.item()["f"]
    nmr_tot_tot_R1 = data.item()["nmr_tot_tot_R1"]
    nmr_tot_tot_R2 = data.item()["nmr_tot_tot_R2"]

    plt.loglog(f, 1/nmr_tot_tot_R1, '.')

    plt.loglog(f, 1/nmr_tot_tot_R2, '.')

    print(ratio, '-',
          np.round(1/np.mean(nmr_tot_tot_R1[1:3])*1e3,2), '-',
          np.round(1/np.mean(nmr_tot_tot_R2[1:3])*1e3,2), '-',
          data.item()["N"])
```
