```python
import numpy as np
import os

import matplotlib.pyplot as plt

datapath = "../../../raw-data/HEWL-in-water/"


all_folders = ["T300K_ratio0.61/", "T300K_ratio2.73/"]

for folder in all_folders:
    if os.path.exists(datapath+folder+"analysed-data.npy"):
        data = np.load(datapath+folder+"analysed-data.npy", allow_pickle=True)
        water_f = data.item()['nmr_water_f']
        water_R1 = data.item()['nmr_water_R1']
        water_R2 = data.item()['nmr_water_R2']
        protein_f = data.item()['nmr_protein_f']
        protein_R1 = data.item()['nmr_protein_R1']
        protein_R2 = data.item()['nmr_protein_R2']

        print("T1-T2-H20", np.round(1/water_R1[0],3), np.round(1/water_R2[0],2))
        print("T1-T2-pro", np.round(1/protein_R1[0],4), np.round(1/protein_R2[0],4))
        print("T1-T2", np.round(1/(water_R1[0]+protein_R1[0]),4), np.round(1/(water_R2[0]+protein_R2[0]),4))

        #plt.loglog(water_f, 1/water_R1, 'ob')
        #plt.loglog(water_f, 1/water_R2, 'or')

        plt.loglog(protein_f, 1/(protein_R1+water_R1), '.b')
        plt.loglog(protein_f, 1/(protein_R2+water_R2), '.r')

        plt.show()
```
